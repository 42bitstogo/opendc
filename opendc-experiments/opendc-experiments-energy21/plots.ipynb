```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pyplot
import seaborn as sns

sns.set()
```

```python
df = pd.read_parquet("output/host-metrics")
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
df
```

```python
x = df.groupby(['power_model', 'host_id', pd.Grouper(freq='1D', key='timestamp')]).mean()
x
```

```python
sns.lineplot(x="timestamp", y="power_draw", hue='power_model', data=x)
```

```python
sns.lineplot(x="timestamp", y="cpu_usage", data=x)
```

```python
sns.lineplot(x="timestamp", y="cpu_demand", data=x)
```

```python
sns.lineplot(x="timestamp", y="requested_burst", data=x)
```

```python
sns.lineplot(x="timestamp", y="granted_burst", data=x)
```

```python
sns.lineplot(x="timestamp", y="overcommissioned_burst", data=x)
```

```python
sns.lineplot(x="timestamp", y="vm_count", data=df.resample('1D').mean())
```
